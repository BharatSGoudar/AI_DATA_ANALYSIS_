```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    'Height': [150, 160, 170, 180, 190],
    'Weight': [50, 60, 70, 80, 90],
    'Size': [10, 15, 13, 18, 12]
})

sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Height', y='Weight', size='Size', sizes=(50, 300), legend=False)
plt.title('Height vs. Weight Scatter Plot with Size Encoding')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.tight_layout()
plt.show()
```
