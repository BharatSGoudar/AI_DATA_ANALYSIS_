```python
# Data Profiling Techniques Examples

# 1. Descriptive Statistics:
# Task 1: Calculate the mean, median, and mode for sales figures in a retail dataset.
# Task 2: Analyze the average age, median, and mode in a customer demographic
# dataset.
# Task 3: Determine the mean, median, and mode of daily website visit counts.
import pandas as pd
data = {'Sales': [200, 220, 250, 200, 270, 220, 230, 210, 250, 260]}
df = pd.DataFrame(data)
mean_sales = df['Sales'].mean()
median_sales = df['Sales'].median()
mode_sales = df['Sales'].mode()[0]
print(f"Mean Sales: {mean_sales}")
print(f"Median Sales: {median_sales}")
print(f"Mode Sales: {mode_sales}")
data = {'Age': [25, 30, 22, 25, 35, 40, 30, 29, 25, 31]}
df = pd.DataFrame(data)
mean_age = df['Age'].mean()
median_age = df['Age'].median()
mode_age = df['Age'].mode()[0]
print(f"Average Age: {mean_age}")
print(f"Median Age: {median_age}")
print(f"Mode Age: {mode_age}")
data = {'Visits': [500, 520, 480, 500, 490, 530, 500, 515, 495, 505]}
df = pd.DataFrame(data)
mean_visits = df['Visits'].mean()
median_visits = df['Visits'].median()
mode_visits = df['Visits'].mode()[0]
print(f"Mean Visits: {mean_visits}")
print(f"Median Visits: {median_visits}")
print(f"Mode Visits: {mode_visits}")






# 2. Distribution Analysis:
# Task 1: Visualize the distribution of product prices using histograms.
# Task 2: Create a distribution chart for exam scores in an educational dataset.
# Task 3: Analyze the distribution of order quantities for inventory management.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example datasets
product_prices = pd.Series([19.99, 29.99, 15.50, 45.00, 22.00, 39.99, 60.00, 29.99, 35.00])
exam_scores = pd.Series([56, 78, 85, 90, 65, 72, 88, 95, 60, 82, 74])
order_quantities = pd.Series([10, 20, 15, 30, 25, 5, 40, 35, 30, 20, 25, 30])

# Set up the plot
plt.figure(figsize=(18, 5))

# Task 1: Product Prices
plt.subplot(1, 3, 1)
sns.histplot(product_prices, bins=10, kde=True, color='skyblue')
plt.title('Product Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)

# Task 2: Exam Scores
plt.subplot(1, 3, 2)
sns.histplot(exam_scores, bins=8, kde=True, color='salmon')
plt.title('Exam Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)

# Task 3: Order Quantities
plt.subplot(1, 3, 3)
sns.histplot(order_quantities, bins=8, kde=True, color='mediumseagreen')
plt.title('Order Quantity Distribution')
plt.xlabel('Quantity Ordered')
plt.ylabel('Frequency')
plt.grid(True)

# Adjust layout and show
plt.tight_layout()
plt.show()







# 3. Data Type Validation:
# Task 1: Validate numeric fields in a dataset to ensure they contain only numbers.
# Task 2: Check for valid date formats in a transaction log.
# Task 3: Validate email formats in a customer contact dataset.
import pandas as pd
import re
from datetime import datetime

# Task 1: Validate numeric fields (e.g., 'price' column)
def validate_numeric_column(df, column):
    if column not in df.columns or df.empty:
        return False
    return pd.to_numeric(df[column], errors='coerce').notna().all()

# Task 2: Validate date formats (e.g., 'transaction_date' column)
def validate_date_column(df, column, date_format='%Y-%m-%d'):
    if column not in df.columns or df.empty:
        return False
    try:
        return df[column].dropna().apply(lambda x: datetime.strptime(x, date_format)).notna().all()
    except ValueError:
        return False

# Task 3: Validate email formats (e.g., 'email' column)
def validate_email_column(df, column):
    if column not in df.columns or df.empty:
        return False
    email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w{2,}$'
    return df[column].dropna().apply(lambda x: bool(re.match(email_pattern, str(x)))).all()








```
