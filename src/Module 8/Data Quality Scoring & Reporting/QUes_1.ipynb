## Read Data from CSV and Calculate DQI

**Description**: Read data from a CSV file, identify errors as missing values, and calculate the DQI.

```python
import pandas as pd
def calculate_dqi(total: int, valid: int, errors: int) -> float:
    if total == 0:
        return 0.0  
    return valid / total
def read_data_and_calculate_dqi(file_path: str, column_name: str) -> float:
    data = pd.read_csv(file_path)
    total_entries = len(data)
    valid_entries = data[column_name].notnull().sum()
    error_entries = total_entries - valid_entries
    dqi = calculate_dqi(total_entries, valid_entries, error_entries)
    return dqi
file_path = '/path/to/your/data.csv'  # Replace with your actual CSV file path
column_name = 'column_of_interest'  # Replace with the column you want to check
try:
    dqi = read_data_and_calculate_dqi(file_path, column_name)
    print(f"Data Quality Index (DQI): {dqi:.2f}")
except FileNotFoundError as e:
    print(f"Error: {e}")
```

### Visualize Basic DQI with Bar Plot

**Description**: Create a bar plot for DQI and errors in a dataset.

```python
#
import matplotlib.pyplot as plt
def plot_dqi_and_errors(dqi, errors):
    categories = ['DQI', 'Errors']
    values = [dqi, errors]
    plt.figure(figsize=(6, 4))
    bars = plt.bar(categories, values, color=['green', 'red'])
    for bar in bars:
        height = bar.get_height()
        plt.annotate(f'{height:.2f}' if isinstance(height, float) else f'{height}',
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3),  # offset text above bar
                     textcoords='offset points',
                     ha='center', va='bottom')
    plt.title('Data Quality Index (DQI) and Errors')
    plt.ylabel('Value')
    plt.ylim(0, max(values) * 1.2)
    plt.show()
dqi_value = 0.85 
error_count = 15  
plot_dqi_and_errors(dqi_value, error_count)
```
