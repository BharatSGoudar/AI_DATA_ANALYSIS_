### Detect Data Drift using Kolmogorov-Smirnov Test
**Description**: Use statistical tests to detect data drift between two datasets.

```python
# write your code from here
import pandas as pd
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt

# Example synthetic datasets (simulate baseline and current)
baseline = pd.DataFrame({
    "amount": [100, 105, 98, 97, 102, 100, 110, 95, 101, 99]
})

current = pd.DataFrame({
    "amount": [150, 160, 148, 152, 149, 170, 145, 158, 165, 155]
})

# Perform KS test on the 'amount' column
statistic, p_value = ks_2samp(baseline["amount"], current["amount"])

print(f"KS Statistic: {statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("ðŸš¨ Drift Detected: The distributions are significantly different.")
else:
    print("âœ… No Drift Detected: The distributions are statistically similar.")

# Optional: Visualize the distributions
plt.hist(baseline["amount"], alpha=0.5, label="Baseline", bins=10)
plt.hist(current["amount"], alpha=0.5, label="Current", bins=10)
plt.legend()
plt.title("Amount Distribution Comparison")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```
